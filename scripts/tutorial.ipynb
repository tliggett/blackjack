# Using pyjack alpha in jupyter

This tutorial will go through the basics of using pyjack 

```python
# Import relevant packages
import pandas as pd
import matplotlib.pyplot as plt
from pyjack21 import BlackJackTable
```

## BlackJackTable

This class simulates hands of blackjack. The results are stored in a pandas dataframe.

In the alpha all players start with 500 and will bet 10 until they lose all of their money

**Optional Parameters:**
+ hands: number of hands to play. Defaults to 1
+ decks: number of decks to put in the shoe. Defaults to 6
+ player_count: number of players at the table. Defaults to 4

```python
player_count = 6
table = BlackJackTable(hands=100, decks=6, player_count=player_count)
table.run()
blackjack_data = table.df
```

```python
blackjack_data
```

```python
for i in range(player_count):
    plt.plot(blackjack_data[f'payroll_{i}'])
```

```python
# Calculate the average payroll
blackjack_data['payroll_avg'] = blackjack_data['payroll_0']
for i in range(player_count):
    blackjack_data['payroll_avg'] = blackjack_data['payroll_avg'] + blackjack_data[f'payroll_{i}']
blackjack_data['payroll_avg'] = blackjack_data['payroll_avg'] / player_count


plt.plot(blackjack_data['payroll_avg'])
```
